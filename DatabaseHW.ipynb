```python
import pandas as pd
import sqlite3

# 1. Load the data from the uploaded file
df_class = pd.read_csv('2026-05-12-class-data.csv')
print("--- 1. CSV Data Loaded Successfully ---")
print(df_class.head(3)) # Prints the first 3 rows from the file
print("\n" + "="*50 + "\n")


# 2. Connect to a local SQLite database file in Colab
conn = sqlite3.connect('class_info.db')
cursor = conn.cursor()

# 3. Create the table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS students (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        first_name TEXT,
        last_name TEXT,
        age INTEGER,
        favorite_country TEXT,
        favorite_food TEXT
    )
''')
conn.commit()


# 4. Loop through and insert the data
for index, row in df_class.iterrows():
    cursor.execute('''
        INSERT INTO students (first_name, last_name, age, favorite_country, favorite_food)
        VALUES (?, ?, ?, ?, ?)
    ''', (row['first_name'], row['last_name'], int(row['age']), row['fav_country'], row['fav_food']))

conn.commit()
print("--- 2. Database Built and Data Populated! ---")
print("\n" + "="*50 + "\n")


# 5. DEMONSTRATE DATA ACCESS (Part 3 of your Homework)
print("--- 3. Homework Demo: Accessing Data from SQLite ---")

# Run a SELECT query to grab the data back out of the database
cursor.execute("SELECT * FROM students")
all_students = cursor.fetchall()

# Loop through the rows from the database and print them cleanly
for student in all_students:
    print(f"Student ID: {student[0]} | Name: {student[1]} {student[2]} | Age: {student[3]} | Prefers: {student[4]} | Food: {student[5]}")


# 6. Clean up the connection
cursor.close()
conn.close()
```
